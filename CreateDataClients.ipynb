```python
import requests
import pandas as pd
```

```python
url_client = "https://faas-lon1-917a94a7.doserverless.co/api/v1/web/fn-a1f52b59-3551-477f-b8f3-de612fbf2769/default/clients-data"
```

```python
path = "transactions_data.csv"
```

```python

df = pd.read_csv(path)
```

```python
unique_clients = df["client_id"].unique()
```

```python
import os
import json,time
# mkdir data
if not os.path.exists("data_clients"):
    os.mkdir("data_clients")
    
    

maxiter = len(unique_clients)
iter = 0

found = False
client_id_not_found = []
for client_id in unique_clients:

    if str(client_id)+".json" in os.listdir("data_clients"):
        print(f"Client {client_id} already downloaded")
        iter += 1
        print(f"{iter}/{maxiter}")
        continue


    response = requests.get(url_client, params={"client_id": client_id})   
    data = response.json()


    if "error" in data.keys():
        print(f"Error in client {client_id}")
    else:
        found = True

    if not found:

        for att in range(10):
            response = requests.get(url_client, params={"client_id": client_id})   
            data = response.json()
            if "error" in data.keys():
                print(f"Error in client {client_id}")
            else:
                found = True
                break

            time.sleep(2)

    if not found:
        client_id_not_found.append(client_id)
        print(f"Client {client_id} not found")
        continue

    with open(f"data_clients/{client_id}.json", "w") as f:
        json.dump(data, f)
    print(f"Client {client_id} data received")
    iter += 1
    print(f"{iter}/{maxiter}")
```
