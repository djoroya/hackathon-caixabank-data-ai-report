```python
import pandas as pd

df = pd.read_csv(path)
```

```python
unique_clients = df["client_id"].unique()
```

```python
import os
import json
import time
# mkdir data
if not os.path.exists("data_cards"):
    os.makedirs("data_cards")


maxiter = len(unique_clients)
iter = 0
client_id_not_found = []
for client_id in unique_clients:

    found = False

    response = requests.get(url, params={"client_id": client_id})   
    data = response.json()

    if "error" in data.keys():
        print(f"Error in client {client_id}")
    else:
        found = True


    if not found:

        for att in range(10):
            response = requests.get(url, params={"client_id": client_id})   
            data = response.json()
            if "error" in data.keys():
                print(f"Error in client {client_id}")
            else:
                found = True
                break

            time.sleep(2)

    if not found:
        client_id_not_found.append(client_id)
        print(f"Client {client_id} not found")
        continue


    with open(f"data_cards/{client_id}.json", "w") as f:
        json.dump(data, f)
    # temp json 
    with open(f"card.temp.json", "w") as f:
        json.dump(data, f)
    print(f"CArd {client_id} data received")
    iter += 1
    print(f"{iter}/{maxiter}")

    time.sleep(2)
    
```
