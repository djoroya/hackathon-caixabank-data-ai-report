```python
path = "data/transactions_data.csv"
path_client = "data_clients.csv"
path_card = "df_cards.csv"
```

```python
import pandas as pd

df = pd.read_csv(path)
```

```python
df_cards = pd.read_csv(path_card)
```

```python
df_cards_visa = df_cards[df_cards["card_brand"] == "Visa"]["card_id"].values
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
import datetime
```

```python
df_Feb_2016 = df[(df['date'] >= datetime.datetime(2016, 2, 1)) & (df['date'] <= datetime.datetime(2016, 2, 29))]
```

```python

df_client = pd.read_csv(path_client)
```

```python
# client age>40

df_client = df_client[df_client['age']>40]
id_client = df_client['client_id'].tolist()
```

```python
df_Feb_2016_Age40 = df_Feb_2016[df_Feb_2016['client_id'].isin(id_client)]
```

```python
df_Feb_2016_Age40
```

```python
df_Feb_2016_Age40_Visa = df_Feb_2016_Age40[df_Feb_2016_Age40['card_id'].isin(df_cards_visa)]
```

```python
df_Feb_2016_Age40_Visa
```

```python
# count of transactions by client_id
df_Feb_2016_Age40_count = df_Feb_2016_Age40_Visa.groupby('client_id').size().reset_index(name='count')
```

```python
# count of transactions by client_id and card_id
df_Feb_2016_Age40_count = df_Feb_2016_Age40_Visa.groupby(['client_id', 'card_id']).size().reset_index(name='count')
```

```python
# date -> datetime
df["date"] = pd.to_datetime(df["date"])
df["amount"] = df["amount"].str.replace("$", "").astype(float)
```

```python
# select 31.12
df_selected = df[df["merchant_city"] == "ONLINE"]
df_selected = df_selected[df_selected["date"].dt.month == 12]
df_selected = df_selected[df_selected["date"].dt.day == 31]
df_selected["amount_abs"] = df_selected["amount"].abs()
df_selected = df_selected.sort_values("amount_abs", ascending=False)
```
