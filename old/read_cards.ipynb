```python
import os,glob,json
```

```python
data_clients = glob.glob("data_cards/*")
```

```python
cards_clients = [ json.load(open(f)) for f in data_clients ]

```

```python
client_id_list = []
card_id_list = []
expire_date_list = []
credit_limit_list = []
card_brand_list = []
for cards_client in cards_clients:
    values = cards_client['values']

    card_ids = values.keys()

    for icard in card_ids:
        card_id_list.append(icard)
        client_id_list.append(cards_client['client_id'])
        expire_date_list.append(values[icard]['expires'])
        credit_limit_list.append(values[icard]["credit_limit"])
        card_brand_list.append(values[icard]["card_brand"])

    
```

```python
import pandas as pd
```

```python
df_expires = pd.DataFrame({'client_id':client_id_list,
                           'card_id':card_id_list,'expire_date':expire_date_list,
                            'credit_limit':credit_limit_list,'card_brand':card_brand_list})
# expires date MM/YYYY 
df_expires['expire_date'] = pd.to_datetime(df_expires['expire_date'],format='%m/%Y')
# credit_limit $ format to float
df_expires['credit_limit'] = df_expires['credit_limit'].str.replace("$","").astype(float)
```

```python
df_expires.to_csv("df_cards.csv",index=False)
```
